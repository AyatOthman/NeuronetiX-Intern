# Introduction

This project involves analyzing a dataset from Kaggle using Multiple Linear Regression. The primary objective of this analysis is to understand how multiple independent variables influence a single dependent variable. By applying Multiple Linear Regression, we aim to build a predictive model that can accurately forecast the target variable based on the input features.

Multiple Linear Regression is an extension of simple linear regression that considers multiple factors simultaneously, allowing us to assess the combined effect of various predictors on the outcome. This analysis will help us uncover relationships, make predictions, and gain insights into the significance of each variable in the dataset.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('multiple_linear_regression_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Exploratory Data Analysis (EDA)

```python
#Correlation Matrix and Heatmap
corr = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Pair Plot
sns.pairplot(df)
plt.suptitle('Pair Plot of the Dataset', y=1.02)
plt.show()
```

```python
# Box plot for detecting outliers in each column
plt.figure(figsize=(12, 6))

# Box plot for the 'age' column
plt.subplot(1, 3, 1)
sns.boxplot(y=df['age'])
plt.title('Box Plot of Age')

# Box plot for the 'experience' column
plt.subplot(1, 3, 2)
sns.boxplot(y=df['experience'], color='orange')
plt.title('Box Plot of Experience')

# Box plot for the 'income' column
plt.subplot(1, 3, 3)
sns.boxplot(y=df['income'], color='green')
plt.title('Box Plot of Income')

plt.show()
```

```python
X = df[['age', 'experience']]  # Features
y = df['income']               # Target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

# Model Selection and Training

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

# Model Evaluation

```python
# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')

# R-squared Score
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2:.2f}')
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Create figure and axes
plt.figure(figsize=(12, 8))

# Plot Actual vs Predicted for Age
plt.subplot(1, 2, 1)  # Create subplot for Age
plt.scatter(X_test['age'], y_test, color='blue', label='Actual')
plt.scatter(X_test['age'], y_pred, color='red', label='Predicted', alpha=0.5)
plt.title('Actual vs Predicted Income (Age)')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()

# Plot Actual vs Predicted for Experience
plt.subplot(1, 2, 2)  # Create subplot for Experience
plt.scatter(X_test['experience'], y_test, color='blue', label='Actual')
plt.scatter(X_test['experience'], y_pred, color='red', label='Predicted', alpha=0.5)
plt.title('Actual vs Predicted Income (Experience)')
plt.xlabel('Experience')
plt.ylabel('Income')
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```

# Summary

The Multiple Linear Regression model effectively captured the relationship between age, experience, and income, with outstanding performance metrics. The high R² value indicates that the model explains almost all of the variability in income, and the reduced MSE signifies accurate predictions.
